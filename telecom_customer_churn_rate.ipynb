```python
import pandas as pd
```

```python
#Loading 1st dataset
df1 = pd.read_csv("churn-bigml-20.csv")
print("df1 loaded")
```

```python
#Loading second dataset
df2 = pd.read_csv("churn-bigml-80.csv")
print("df2 loaded")
```

```python
# Combining the two datasets by stacking them on top of each other so that I can have 1 large data set.
combined_df = pd.concat([df1, df2], axis=0, ignore_index=True)
print("Datasets combined")
```

```python
#saving the combined dataset
combined_df.to_csv('telecom_customer_data', index=False)
```
